```python
import openai
from datetime import datetime

class DeepSeekConversation:
    def __init__(self, api_key):
        self.client = openai.OpenAI(
            api_key=api_key,
            base_url="https://api.deepseek.com/v1"
        )
        self.reset_conversation()
        
    def reset_conversation(self):
        self.messages = [
            {"role": "system", "content": 
             "You are a world-class empathetic matchmaker and clinical psychologist."
             "Extract the user's subconscious values and attachment style through organic conversation"
             "Only ask a single question in each of your text output"
             "Ask 3 to 5 meaningful questions in total"
             "After you think you understand the user, you will give out 2 summaries"
             "Summary 1 will be about what kind of person the user is, Summary 2 will be about the kind of person that the user desires"
            }
        ]
        print("✓ New conversation started")
            
            
        # self.conversation_history = []

    def chat(self, user_input):
        # Add to conversation history
        # self.conversation_history.append({
        #     "timestamp": datetime.now().strftime("%H:%M:%S"),
        #     "role": "user",
        #     "content": user_input
        # })
        
        # Add to messages for API
        self.messages.append({"role": "user", "timestamp": datetime.now().strftime("%H:%M:%S"), "content": user_input})
        
        try:
            response = self.client.chat.completions.create(
                model="deepseek-chat",
                messages=self.messages,
                temperature=0.7,
                max_tokens=1000
            )
            
            assistant_reply = response.choices[0].message.content
            
            # # Store in history
            # self.conversation_history.append({
            #     "timestamp": datetime.now().strftime("%H:%M:%S"),
            #     "role": "assistant",
            #     "content": assistant_reply
            # })
            
            # Add to messages for context
            self.messages.append({"role": "assistant", "content": assistant_reply})
            
            return assistant_reply
            
        except Exception as e:
            return f"❌ Error: {e}"
    
    def show_history(self):
        print("\n" + "="*50)
        print("CONVERSATION HISTORY")
        print("="*50)
        for entry in self.messages:
            print(f"[{entry['timestamp']}] {entry['role'].upper()}: {entry['content'][:100]}...")
        print("="*50)
    
    def get_context_length(self):
        total_tokens = sum(len(msg["content"]) for msg in self.messages)
        print(f"Current context: {len(self.messages)} messages, approx {total_tokens} characters")
```

```python
chat = DeepSeekConversation("sk-032be710b2254d8f81939277759eb423")

# Interactive loop
print("\n🤖 DeepSeek Interactive Chat")
print("Commands: 'history' - show conversation, 'context' - show context size, 'reset' - clear, 'quit' - exit")
print("-" * 50)

response = chat.chat("")
print(f"\n🤖 DeepSeek: {response}")

while True:
    user_input = input("\n👤 You: ")
    print(f"\n🧔 You: {user_input}")
    
    if user_input.lower() in ['quit', 'exit', 'q']:
        print("\n👋 Goodbye!")
        break
    elif user_input.lower() == 'reset':
        chat.reset_conversation()
        continue
    elif user_input.lower() == 'history':
        chat.show_history()
        continue
    elif user_input.lower() == 'context':
        chat.get_context_length()
        continue
    response = chat.chat(user_input)
    print(f"\n🤖 DeepSeek: {response}")
```

```python
from sentence_transformers import SentenceTransformer, util

# 1. Load a pre-trained model (No training required!)
# This model turns text into a 384-dimensional vector
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
import json
file_path = 'data.json'

# Use a 'with' statement to open and automatically close the file
with open(file_path, 'r') as file:
    data = json.load(file)

print(data)
```

```python
candidate_vectors = []
for i in data:
    candidate_vectors.append({
        "name": i['name'],
        "bio_vector": model.encode(i['bio']),
        "desired_vector": model.encode(i['desired_type'])
    })
```

```python
all_match_scores = []
for i in range(20):
    for j in range(i+1, 20):
        cosine_scores1 = util.cos_sim(candidate_vectors[i]['bio_vector'], candidate_vectors[j]['desired_vector'])[0]
        cosine_scores2 = util.cos_sim(candidate_vectors[j]['bio_vector'], candidate_vectors[i]['desired_vector'])[0]
        all_match_scores.append({
            "A": candidate_vectors[i]["name"],
            "B": candidate_vectors[j]["name"],
            "A是B的理想型": cosine_scores1,
            "B是A的理想型": cosine_scores2,
            "平均分": ( cosine_scores1 + cosine_scores2 ) / 2
        })

```
